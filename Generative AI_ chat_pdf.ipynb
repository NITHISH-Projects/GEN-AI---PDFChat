````python
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import CohereEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import ChatVectorDBChain
from langchain.llms import Cohere
import cohere
import streamlit as st
````

#Loding Document into vector database

````python
pdf_path = "./Feature Guide.pdf"
loader = PyPDFLoader(pdf_path)
pages = loader.load_and_split()
print(pages[0].page_content)
````

#Converting to Embeddings and persisting vector database

Cohere API Generation :
1. Login to https://dashboard.cohere.com/welcome/login
2. Sign up -> preferred Google
3. Fill up the details
4. Navigate to API Keys section
5. View and copy the API key 
6. Replace the COHERE_API_KEY with this API key obtained

````python
COHERE_API_KEY = <Your_Cohere_api_key>

embeddings = CohereEmbeddings(cohere_api_key=COHERE_API_KEY, user_agent="my-app")
vectordb = Chroma.from_documents(pages, embedding=embeddings, persist_directory=".")

vectordb.persist()
````

#Chaining LLM, Vector Database, User question and answer with grounded response.

````python

pdf_qa = ChatVectorDBChain.from_llm(
    Cohere(cohere_api_key=COHERE_API_KEY),
    vectordb,
    return_source_documents=True,
)

query = input("Please Enter your question: ")

#write code after this
result = pdf_qa({"question": query, "chat_history":""})


print(result)

````
